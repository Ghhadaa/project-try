## 1) Introduction

A digital **grayscale** image can be represented as a **2D array** (matrix) where each entry is a pixel intensity:

- 0 → black  
- 255 → white  
- values in between → shades of gray  

Mathematically, we write:

\[ I(x,y) \in [0,255] \]

Because images are matrices, we can perform **mathematical operations** on them (often **element-wise**):
- Image + image
- Image − image
- Image × scalar (contrast)
- Image + scalar (brightness)

> Important: Pixel values must stay within a valid range. For 8-bit images, use **clipping** to keep values in \([0,255]\).


## 2) Setup: Import Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import data

# Helper function for display
def show(img, title="", cmap="gray"):
    plt.imshow(img, cmap=cmap)
    plt.title(title)
    plt.axis('off')
    plt.show()
#  hiiiiiiii !!!

```

## 3) Load a Sample Image

We'll use a built-in grayscale image from `skimage`.


```python
img = data.camera()   # grayscale image (2D array)
show(img, "Original Image")

print("Shape:", img.shape)
print("Data type:", img.dtype)
print("Min/Max:", img.min(), img.max()) 
```

## 4) Image as a Matrix (Concept)

An image is a matrix like:

\[
I =
\begin{bmatrix}
120 & 80 & 200 \\
30 & 150 & 90 \\
255 & 60 & 10
\end{bmatrix}
\]

Each number is a pixel intensity.


## 5) Image Addition

### A) Image + Image (element-wise)

\[ C(x,y) = A(x,y) + B(x,y) \]

To add two images, they must have the **same shape**.


```python
img2 = np.copy(img)
addition = img + img2

# Display the result directly (notice any artifacts?)
show(addition, "Image + Image (no clipping)")

print("Result dtype:", addition.dtype)
print("Min/Max:", addition.min(), addition.max())
```

## 6) Clipping (Very Important)

Pixel intensities must stay in **[0,255]** for 8-bit images.

If values exceed 255 (or go below 0), results can overflow/underflow and look wrong.

Use:

\[ \text{clip}(I, 0, 255) \]


```python
addition_clip = np.clip(img.astype(np.float32) + img2.astype(np.float32), 0, 255).astype(np.uint8)
show(addition_clip, "Image + Image (with clipping)")

print("Min/Max after clipping:", addition_clip.min(), addition_clip.max())
```

## 7) Scalar Addition (Brightness Increase)

\[ C(x,y) = I(x,y) + k \]

Adding a positive constant makes the image **brighter**.


```python
bright = np.clip(img.astype(np.float32) + 50, 0, 255).astype(np.uint8)
show(bright, "Brightness Increased (+50)")
```

## 8) Subtraction

\[ C(x,y) = A(x,y) - B(x,y) \]

Subtraction can highlight differences.


```python
sub = np.clip(img.astype(np.float32) - img2.astype(np.float32), 0, 255).astype(np.uint8)
show(sub, "Subtraction (A - B) with clipping")
```

## 9) Scalar Multiplication (Contrast)

\[ C(x,y) = \alpha I(x,y) \]

- If \(\alpha > 1\): increases contrast (brighter highlights)
- If \(0 < \alpha < 1\): decreases contrast (darker/less intense)


```python
contrast_low = np.clip(0.6 * img.astype(np.float32), 0, 255).astype(np.uint8)
contrast_high = np.clip(1.6 * img.astype(np.float32), 0, 255).astype(np.uint8)

show(contrast_low, "Contrast: 0.6 × Image")
show(contrast_high, "Contrast: 1.6 × Image")
```
